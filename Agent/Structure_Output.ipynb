```python
from langchain_openai import ChatOpenAI
import os
from dotenv import load_dotenv
load_dotenv()

llm = ChatOpenAI(model_name="gpt-5-mini")
```

```python
response = llm.invoke("What is the capital of France? generate the answer in format of \"Q: <question> A: <answer>\"")
response = response.content
print(response)
```

## Pydantic

```python
from pydantic import BaseModel,Field

class llm_schema(BaseModel):
    question: str = Field(description="The question asked")
    answer: str = Field(description="The answer to the question")
    token_usage: int = Field(description="The number of tokens used to generate the answer")
```

```python
llm_structured = llm.with_structured_output(llm_schema)
response = llm_structured.invoke("What is the capital of France?")
response
```
